*Summary*: This dataset contains 16 years of Netflix movie records, spanning from 2008 up to 2023. It includes detailed information about movies available onn Netflix such as title, director, release year, genre, duration, country of production, and ratings.

*Objective*: The objective of this project is to perform an in-depth exploratory data analysis (EDA) of Netflix movies to discover significant patterns and insights. The analysis will emphasize genres, countries, languages, and directors to determine which categories are most popular and best rated. In addition, broader metrics such as ratings, popularity, budget, and revenue will be explored to identify performance trends. This approach will enable a comparative view of different aspects of movies and highlight the key factors contributing to audience engagement and commercial success.

### Country Analysis

In this section, we will analyze Netflix movies based on their production countries. The
goal is to:

- Identify the top countries producing the most content.
- Compare countries by average ratings, budget and revenue.
- Analyze average rating and average popularity of countries.
- Study the trend of movie production over the years for leading countries.

```python
# country wise explode with primary country and equal profit distribution

# Split and explode country column into list
df['country'] = df['country'].str.split(', ')
df_exploded = df.explode('country')
# create primary country
df['primary_country'] = df['country'].apply(lambda x: x[0].strip() if isinstance(x, list) else x)

# Remove extra spaces
df_exploded['country'] = df_exploded['country'].astype(str).str.strip()

# Divide profit equally among countries
df_exploded['profit_divided'] = df_exploded['profit'] / df_exploded.groupby('show_id')['country'].transform('count')
```

```python
# Check the DataFrame after exploding and add primary column
df_exploded.iloc[:,[5,17]].head(5)
```

```python
# Group by 'country' to get count and average metrics for easier country-wise analysis

countries_group = (df_exploded[df_exploded['country'] != 'Unknown'].groupby('country').agg(
    number_of_movies = ('show_id', 'count'),
    avg_rating = ('rating', 'mean'),
    avg_popularity = ('popularity', 'mean'),
    avg_vote_count = ('vote_count', 'mean'),
    avg_revenue = ('revenue', 'mean'),
    avg_budget = ('budget', 'mean'),
    avg_profit = ('profit_divided', 'mean')
    ).reset_index().sort_values(by = 'number_of_movies', ascending=False))

# Filter countries with at least 500 movies
filtered_countries = countries_group[countries_group['number_of_movies'] >= 500]
```

### Top Producing Countries

```python
sns.barplot(x='number_of_movies', y='country', data=filtered_countries.head(10))
plt.xlabel('Number of Movies')
plt.ylabel('Country')
plt.title('Top Producing Countries on Netflix')
plt.show()
```

#### Compare Budget, Revenue and Profit by Countries

```python
from matplotlib.ticker import FuncFormatter
def millions(x, pos):
    return f'{x/1_000_000:.1f}M'

# Budget Plot
fig, ax = plt.subplots(figsize=(16, 8))
sns.barplot(
    x='avg_budget', 
    y='country', 
    data=filtered_countries.sort_values(by='avg_budget', ascending=False).head(10)
)
ax.set_title('Top 10 Countries by Average Budget')
ax.set_xlabel('Budget (log scale)')
ax.set_ylabel('Country')
ax.xaxis.set_major_formatter(FuncFormatter(millions))
plt.tight_layout()
plt.show()
```

```python
# Revenue Plot
fig, ax = plt.subplots(figsize=(16, 8))
sns.barplot(
    x='avg_revenue',
    y='country',
    data=filtered_countries.sort_values(by='avg_revenue', ascending=False).head(10),
    ax=ax
)
ax.set_title('Top 10 Countries by Average Revenue')
ax.set_xlabel('Revenue (log scale)')
ax.set_ylabel('Country')
ax.xaxis.set_major_formatter(FuncFormatter(millions))
plt.tight_layout()
plt.show()
```

```python
# Profit Plot
fig, ax = plt.subplots(figsize=(16, 8))
sns.barplot(
    x='avg_profit',
    y='country',
    data=filtered_countries.sort_values(by='avg_profit', ascending=False).head(10),
    ax=ax
)
ax.set_title('Top 10 Countries by Average Profit')
ax.set_xlabel('Profit (log scale)')
ax.set_ylabel('Country')
ax.xaxis.set_major_formatter(FuncFormatter(millions))
plt.tight_layout()
plt.show()
```

### Top Average Rating Countries

```python
# Average Rating Countries
sns.barplot(
	x='avg_rating',
	y='country',
	data=filtered_countries.sort_values(by='avg_rating', ascending=False).head(10)
)
plt.xlabel("Average Rating")
plt.ylabel("Country")
plt.title("Top 10 Countries by Average Rating")
plt.show()
```

### Top Average Popularity Countries

#### Movie Release Trend of Top 5 Countries

```python
# Top 5 countries
top5_countries = df_exploded.groupby('country')['show_id'].count().sort_values(ascending=False).head(5).index.tolist()

# filter
df_top5 = df_exploded[df_exploded['country'].isin(top5_countries)]

# Group by
trend_df = df_top5.groupby(['country', 'release_year'], as_index=False)['show_id'].count()

# plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='release_year', y='show_id', hue='country', data=trend_df, marker='o')
plt.title('Movie Release per Year of Top 5 Countries')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Countries')
plt.show()
```

### Language Analysis

In this section, we analyze Netflix movies based on languages. The goal is to identify:
- Most common languages used in Netflix movies.
- Languages associated with the highest average ratings.
- Languages generating the most profit.

#### Language-wise Movie Distribution

```python
# Top 5 languages by number of movies
top_languages = df['language'].value_counts().head(5)

# Calculate 'Others' as sum of remaining languages
others_count = df['language'].value_counts().iloc[5:].sum()
top_languages['Others'] = others_count

# Pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    top_languages,
    labels=top_languages.index,
    autopct='%1.1f%%',
    startangle=140,
)

# Set Language Labels to Black
for text in texts:
    text.set_color('Black')

# Set Percentage text to White
for autotext in autotexts:
    autotext.set_color('White')

plt.title("Movie Distribution by Language", fontsize=14, fontweight='bold', color='blue')
plt.show()
```

```python
# Group by 'language' to get count and average metrics for easier language wise analysis
language_group = (
    df_exploded.groupby('language')
    .agg(
        number_of_movies = ('show_id', 'count'),
        avg_rating = ('rating', 'mean'),
        avg_popularity = ('popularity', 'mean'),
        avg_vote_count = ('vote_count', 'mean'),
        avg_revenue = ('revenue', 'mean'),
        avg_budget = ('budget', 'mean'),
        avg_profit = ('profit', 'mean')
    ).reset_index().sort_values(by='number_of_movies', ascending=False)
)
```

#### Top Average Rating Languages

```python
# Plot Average Rating Languages
sns.barplot(
    x='avg_rating',
    y='language',
    data=language_group.sort_values(by='avg_rating', ascending=False).head(10)
)
plt.xlabel("Avgerage Rating")
plt.ylabel("Language")
plt.title("average Rating by Language")
plt.show()
```

#### Top Average Porfit Languages

```python
# plot of Average Profit Languages
sns.barplot(
    x='avg_profit',
    y='language',
    data=language_group.sort_values(by='avg_profit', ascending=False).head(10)
)
plt.xlabel("Average Profit")
plt.ylabel("Language")
plt.title("Average Profit by Language")
ax = plt.gca()
ax.xaxis.set_major_formatter(FuncFormatter(millions))
plt.show()
```

### Director Analysis

In this section, we analyze Netflix movies based on directors. The goal is to identify:
- Top directors with the highest number of movies.
- Most profitable directors.
- Directors with the best average ratings.

#### Top directors with the highest number of movies.

```python
# Director wise grouping
director_group = (
    df_exploded[df_exploded['director'] != 'Unknown'].groupby('director')
    .agg(
        number_of_movies = ('show_id', 'count'),
        avg_rating = ('rating', 'mean'),
        avg_popularity = ('popularity', 'mean'),
        avg_vote_count = ('vote_count', 'mean'),
        avg_revenue = ('revenue', 'mean'),
        avg_budget = ('budget', 'mean'),
        avg_profit = ('profit', 'mean'),
    ).reset_index().sort_values(by='number_of_movies', ascending=False)
)

# Plot the top 10 directors with the highest number of movies using a horizontal bar chart
sns.barplot(
    x='number_of_movies',
    y='director',
    data=director_group.head(10)
)
plt.title('Number of Movies Directed by Each Director')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
plt.show()
```

#### Directors with the best average ratings.

```python
# Directors with more than 5 movies to ensure reliable averages
filtered = director_group[director_group['number_of_movies'] > 5]

# Plot Average Rating by Director
sns.barplot(
    x='avg_rating',
    y='director',
    data=filtered.sort_values(by='avg_rating', ascending=False).head(10))
plt.title("Top 10 Directors with Highest Average Ratings")
plt.xlabel("Average Rating", fontsize=12)
plt.ylabel("Director", fontsize=12)
plt.show()
```

#### Top Profitable Directors

```python
# Select the Top 10 directors with the highest average profit
top_profitable = filtered.sort_values(by='avg_profit', ascending=False).head(10)

# Bar plot of the top 10 directors
sns.barplot(
    x='avg_profit',
    y='director',
    data=top_profitable
)
plt.title("Top 10 Most Profitable Directors")
plt.xlabel("Average Profit", fontsize=12)
plt.ylabel("Director", fontsize=12)
ax = plt.gca()
ax.xaxis.set_major_formatter(FuncFormatter(millions))
plt.show()
```
